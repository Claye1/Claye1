# **TOP 20 BASED ON TRANSACTIONS**

```python
import pandas as pd

# Read data from the file, assuming it's a CSV
df = pd.read_excel('/content/drive/MyDrive/DEMOLYTICS PROJECT/New XeroData - Leesi 2/XeroData - LEESI 2.xlsx')

# Group by 'SICCode_Industry' and count the number of transactions
SICCode_transaction_counts = df.groupby('SICCode_Industry').size()

# Sort the counts in descending order and get the top 20
top_10_siccodes = SICCode_transaction_counts.sort_values(ascending=False).head(20)

# Display the top 10 ICCode with the highest number of transactions
print(top_10_siccodes)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Group by 'SICCode' and count the number of transactions
iccode_transaction_counts = df.groupby('SICCode_Industry').size()

# Sort the counts in descending order and get the top 10
top_10_iccodes = iccode_transaction_counts.sort_values(ascending=False).head(20)

# Set up a color palette
colors = sns.color_palette('husl', len(top_10_iccodes))  # 'husl' gives distinct colors

# Plot the top 10 ICCode values based on the number of transactions as a horizontal bar chart
plt.figure(figsize=(14, 8))
top_10_iccodes.plot(kind='barh', color=colors)

# Add labels and title
plt.xlabel('Number of Transactions')
plt.ylabel('SICCode_Industry')
plt.title('Top 20 ICCode in Industry Based on Transactions')

# Invert the y-axis to have the highest value at the top
plt.gca().invert_yaxis()

# Show the plot
plt.show()
```

# **BOTTOM 10 INDUSTRIES BASED ON TRANSACTIONS**

```python
import pandas as pd

# Read data from the file, assuming it's a CSV
df = pd.read_excel('/content/drive/MyDrive/DEMOLYTICS PROJECT/New XeroData - Leesi 2/XeroData - LEESI 2.xlsx')

# Group by 'SICCode_Industry' and count the number of transactions
SICCode_transaction_counts = df.groupby('SICCode_Industry').size()

# Sort the counts in descending order and get the top 20
bottom_10_siccodes = SICCode_transaction_counts.sort_values(ascending=False).tail(10)

# Display the top 10 ICCode with the highest number of transactions
print(bottom_10_siccodes)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Group by 'SICCode_Industry' and count the number of transactions
SICCode_transaction_counts = df.groupby('SICCode_Industry').size()

# Sort the counts in descending order and get the bottom 10
bottom_10_siccodes = SICCode_transaction_counts.sort_values(ascending=False).tail(10)

# Set up a color palette for distinct colors
colors = sns.color_palette('husl', len(bottom_10_siccodes))  # 'husl' gives distinct colors

# Plot the bottom 10 SICCode_Industry values as a horizontal bar chart
plt.figure(figsize=(12, 8))
bottom_10_siccodes.plot(kind='barh', color=colors)

# Invert the y-axis to have the highest value at the top
plt.gca().invert_yaxis()

# Add labels and title
plt.xlabel('Number of Transactions')
plt.ylabel('SICCode_Industry')
plt.title('Bottom 10 SICCode_Industry Based on Number of Transactions')

# Show the plot
plt.show()
```
